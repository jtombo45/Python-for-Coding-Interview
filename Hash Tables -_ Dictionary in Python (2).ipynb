#### Problem 4: Given a list of names, write a function that finds a name that appears twice in the list.
For examples, if you're given["George","Tom","Emily","Jenny","Tom"], your function should return Tom

```python
#I: list
#O: string
#E: list that contains only one character
#A: One unique solution -> only one string appears 2

def appearTwice(arr):
    dt = {}
    #Iterating through the list of names
    for element in arr:
        #print("element: " + str(element))
        if element not in dt:
            #print("hi")
            dt[element] = 1 #0 is the value which represent the appeareance of the name
        else:
            dt[element] += 1 #increasing the number of appearance
    #print("\n" + str(dt))
    for element in dt:
        if dt[element] == 2:
            return element
    return ""

#O(n)+O(n) T -> O(n)
```

```python
arr = ["George","Tom","Emily","Jenny","Tom"]

print("The name that appears twice is " + str(appearTwice(arr)))
```

theTEWEngineeringWorld Solution:
- Just assume that there's only one name that appears twice

```python
def appear_twice(arr):
    dt = {} #stores the name and the # of times the names appear
    for name in arr:
        #print(name)
        if name in dt:
            return name
        else:
            dt[name] = 1
    return ""
```

```python
arr = ["George","Tom","Emily","Jenny","Tom"]
print("The name that appears twice is "+ str(appear_twice(arr)))
```

His solution work only if we can assume that one name can only appear twice and the rest of the names only appears once
His solution doesn't work for cases where the names can appear more than twice

```python
def sumOfTen(arr):
    #key and value
    pair = {}
    
    #Iterate through arr and compare each number to each other:
    for i in range(len(arr)):
        for j in range(len(arr)):
            #If numbers equal to 10:
            if i != j and arr[i]+arr[j] == 10:
                pair[arr[i]] = arr[j]
    #If a pair is found:
    if len(pair) >= 1:
        fNum = [key for key in pair.keys()][0]
        sNum = [val for val in pair.values()][0]
        print("first number: " + str(fNum))
        print("second number: " + str(sNum))
    else:
        print("Not found")
        
        
    
```
